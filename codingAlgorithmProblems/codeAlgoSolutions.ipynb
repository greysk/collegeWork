```python
def format_duration(seconds):
    if seconds == 0:
        return 'now'
    if seconds < 60:
        return f'{seconds} second{"s" if seconds == 1 else ""}'

    duration = []
    minute = 60
    hour = minute * 60
    day = hour * 24
    year = day * 265
    
    if seconds >= year:
        duration.append(
            f'{seconds // year} year{"s" if seconds >= year * 2 else ""}')
        seconds %= year
    if seconds >= day:
        duration.append(
            f'{seconds // day} day{"s" if seconds >= day * 2 else ""}')
        seconds %= day
    if seconds >= hour:
        duration.append(
            f'{seconds // hour} hour{"s" if seconds >= hour * 2 else ""}')
        seconds %= hour
    if seconds >= minute:
        duration.append(
            f'{seconds // minute} minute{"s" if seconds >= minute * 2 else ""}')
        seconds %= minute
    if seconds:
        duration.append(f'{seconds} second{"s" if seconds == 1 else ""}')

    if len(duration) > 1:
        human_duration = ','.join(duration[0:len(duration-1)])
        return f'{human_duration} and {duration[-1]}'
    return duration[0]

format_duration(1)
```

# Counting Principles

```python
from collections.abc import Sequence

def fundamental_counting_principle(option_nums: Sequence) -> int:
    """Returns option_num[0] * option_num[1] * ... * option_num[-1]"""
    num_possibilities = 0

    for iopt in option_nums:
        num_possibilities *= iopt

    return num_possibilities


def factorial(n: int) -> int:
    """factorial(n) = n * (n-1) * ... * 2 * 1"""
    if n < 0:
        raise NotImplemented

    total_num_permutations = 1

    while n > 1:
        total_num_permutations *= n
        n -= 1

    return total_num_permutations


def permutation(n: int, r: int) -> int:
    """Premutation of n elements taken r times.
    
    Returns factorial(n) / factorial(n - r)
    """
    return factorial(n) / factorial(n - r)


def distinguishable_permutation(option_nums) -> int:
    """Outcome possibilities removing 'duplicate' results.
    
    factorial(n) / (factorial(option_nums[0] * factorial(option_nums[1]) * ... * factorial(option_nums[-1]))
    """
    dividend = 0
    divisor = 1
    for arg in option_nums:
        dividend += arg
        divisor *= factorial(arg)
    return factorial(dividend) / divisor


def combination(n: int, r: int) -> int:
    """Combination of n elements taken r times.
    
    factorial(n) / ((n-r)r!)
    """
    return permutation(n, r) / factorial(r)
```
